### Q. Find the top 10 cuisiones

```python
final_df['Cuisines'].value_counts()
```

```python
final_df[final_df['Aggregate rating']>=4.5].groupby('Cuisines').size().reset_index().head(50)
```

```python
top_cuisines = (
    final_df[final_df['Aggregate rating'] >= 4.5]
    .groupby('Cuisines')
    .size()
    .reset_index(name='Count')
    .sort_values(by='Count', ascending=False)
    .head(10)
)

print(top_cuisines)
```
